```python
import torch 
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np

class UserLatentDiscoverer(nn.Module):
    def __init__(self, user_features_dim, item_latent_dim):
        super(UserLatentDiscoverer, self).__init__()
        self.shared_fc = nn.Sequential(
            nn.Linear(user_features_dim + item_latent_dim, 64),  # Combine user features and item latents
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
        )
        self.predict_fc = nn.Sequential(
            nn.Linear(32, 1),  # Predict score
            nn.Sigmoid()       # Scale output to [0, 1]
        )
        self.latent_fc = nn.Linear(32, item_latent_dim)  # Update user latent traits

    def forward(self, user_features, item_latent):
        x = torch.cat([user_features, item_latent], dim=1)  # Concatenate inputs
        shared_output = self.shared_fc(x)
        predicted_score = self.predict_fc(shared_output)  # Scalar output for score
        updated_user_latent = self.latent_fc(shared_output)  # Updated user latent traits
        return predicted_score, updated_user_latent


class ItemLatentDiscoverer(nn.Module):
    def __init__(self, item_features_dim, user_latent_dim):
        super(ItemLatentDiscoverer, self).__init__()
        self.shared_fc = nn.Sequential(
            nn.Linear(item_features_dim + user_latent_dim, 64),  # Combine item features and user latents
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
        )
        self.predict_fc = nn.Sequential(
            nn.Linear(32, 1),  # Predict score
            nn.Sigmoid()       # Scale output to [0, 1]
        )
        self.latent_fc = nn.Linear(32, user_latent_dim)

    def forward(self, item_features, user_latent):
        x = torch.cat([item_features, user_latent], dim=1)
        shared_output = self.shared_fc(x)
        predicted_score = self.predict_fc(shared_output)
        updated_item_latent = self.latent_fc(shared_output)  
        return predicted_score, updated_item_latent


# Hyperparameters
user_features_dim = 29
item_features_dim = 19
latent_dim = 32  # Size of the latent trait vectors
learning_rate = 0.01
num_epochs = 10

# Load and preprocess data from MovieLens 100K dataset
# Assuming files "u.user", "u.item", and "u.data" are available
users = pd.read_csv('ml-100k/u.user', sep='|', names=['UserID', 'Age', 'Gender', 'Occupation', 'Zip-code'], encoding='latin-1')
items = pd.read_csv('ml-100k/u.item', sep='|', names=['MovieID', 'Title', 'ReleaseDate', 'VideoReleaseDate', 'IMDbURL'] + [f'Genre_{i}' for i in range(19)], encoding='latin-1')
data = pd.read_csv('ml-100k/u.data', sep='\t', names=['UserID', 'MovieID', 'Rating', 'Timestamp'], engine='python')

# Preprocess user features
users['Gender'] = users['Gender'].map({'M': 0, 'F': 1})
age_bins = [0, 18, 25, 35, 45, 50, 56, 100]
users['Age'] = pd.cut(users['Age'], bins=age_bins, labels=False)
user_features = pd.get_dummies(users[['Age', 'Gender', 'Occupation']], columns=['Age', 'Occupation']).astype(np.float32).values

# Preprocess item features
item_features = items.iloc[:, 5:].values  # Use genre information as features

# Merge data for training
merged_data = data.merge(users, on='UserID').merge(items, on='MovieID')
# Normalize user_features and item_features to [0, 1]
user_features = (user_features - user_features.min()) / (user_features.max() - user_features.min())
item_features = (item_features - item_features.min()) / (item_features.max() - item_features.min())

# Extract features and ratings
user_features_data = np.array([user_features[row['UserID'] - 1] for _, row in merged_data.iterrows()])
item_features_data = np.array([item_features[row['MovieID'] - 1] for _, row in merged_data.iterrows()])
ratings = merged_data['Rating'].values

# Convert to tensors
user_features = torch.tensor(user_features_data, dtype=torch.float32)
item_features = torch.tensor(item_features_data, dtype=torch.float32)
ratings = torch.tensor(ratings, dtype=torch.float32).unsqueeze(1)

# Initialize latent traits
user_latent = torch.randn(user_features.size(0), latent_dim, requires_grad=True)
item_latent = torch.randn(item_features.size(0), latent_dim, requires_grad=True)

# Initialize models
ULD = UserLatentDiscoverer(user_features_dim, latent_dim)
ILD = ItemLatentDiscoverer(item_features_dim, latent_dim)

# Initialize user history (assuming it has the same dimensions as user_latent)
user_history = torch.randn(user_features.size(0), latent_dim, requires_grad=True)

# Optimizers
optimizer_ULD = optim.Adam(ULD.parameters(), lr=learning_rate)
optimizer_ILD = optim.Adam(ILD.parameters(), lr=learning_rate)

# Loss function
criterion = nn.HuberLoss()


# Training loop based on the provided diagram
for epoch in range(num_epochs):
    print(f"Epoch {epoch + 1}/{num_epochs}")

    # Step 1: Update user latent traits
    optimizer_ULD.zero_grad()
    predicted_ratings_ULD, updated_user_latent = ULD(user_features, item_latent)
    loss_ULD = criterion(predicted_ratings_ULD, ratings)
    loss_ULD.backward()
    optimizer_ULD.step()

    # Update user_latent based on ULD output
    user_latent = updated_user_latent.detach()

    # Step 2: Update item latent traits
    optimizer_ILD.zero_grad()
    predicted_ratings_ILD, updated_item_latent = ILD(item_features, user_latent)
    loss_ILD = criterion(predicted_ratings_ILD, ratings)
    loss_ILD.backward()
    optimizer_ILD.step()

    # Update item_latent based on ILD output
    item_latent = updated_item_latent.detach()

    print(f"Epoch [{epoch+1}/{num_epochs}] Loss ULD: {loss_ULD.item():.4f}, Loss ILD: {loss_ILD.item():.4f}")

print("Training complete!")

# Example prediction
sample_user_features = user_features[0].unsqueeze(0)
sample_item_latent = item_latent[0].unsqueeze(0)
predicted_rating, _ = ULD(sample_user_features, sample_item_latent)
print(f"Predicted Rating: {predicted_rating.item():.4f}")
```
